## Module 호출

```python
import pandas as pd
import numpy as np
import pickle
```

## Pickle 파일 호출

```python
patent_data = pd.read_pickle('patent_data.pkl')
app_sample = pd.read_pickle('app_sample.pkl')
```

```python
# Add name column(last name + first name)
# Equalize organization shape 
patent_data['name'] = patent_data['name_last'].str.upper() +' '+ patent_data['name_first'].str.upper()
patent_data['organization'] = patent_data['organization'].str.upper()

# name_last and name_first drop & columns reindexing
patent_data = patent_data.drop(['name_last'],axis = 1)
patent_data = patent_data.drop(['name_first'],axis = 1)
patent_data = patent_data.reindex(columns=['inventor_id', 'name', 'organization', 'application_id', 'number_updated','number'])
```

```python
# Erase'\r' in number_updated and rename
patent_data['number'] = patent_data['number_updated'].apply(lambda x : str(x)[:-1])
patent_data = patent_data.drop(['number_updated'],axis = 1)
```

```python
# merge patent_data and app_sample on number
patent = pd.merge(patent_data, app_sample, on=['number'])
```

## Pickle 파일 호출

```python
citation =  pd.read_pickle('citation_list(no_set).pkl')
```

```python
# Erase both side blank patent_id data 
citation['patent_id'] = citation['patent_id'].map(lambda x : str(x).strip())

# Merge patnet and citation tables on patent_id
citation_list = pd.merge(patent, citation, on = ['patent_id'])
```

```python
#app_id_x : 저자가 쓴 논문 app_id
#number_x : 저자가 쓴 논문 num
#patent_id : 저자가 쓴 논문 patent_id
#app_id_y : 인용한 논문 app_id
#number_y : 인용한 논문 num
citation_list
```

```python
# rename column's names and drop some columns
citation_list['application'] = citation_list['application_id_x']
citation_list['citation_app'] = citation_list['application_id_y']

citation_list = citation_list.drop(['application_id_x'],axis = 1)
citation_list = citation_list.drop(['number_x'],axis = 1)
citation_list = citation_list.drop(['application_id_y'],axis = 1)
citation_list = citation_list.drop(['number_y'],axis = 1)
```

## Pickle 파일 호출

```python
raw_inventor = pd.read_pickle('./raw_inventor.pkl')
```

```python
# rename column's names and drop some columns
raw_inventor['citation_app'] = raw_inventor['application_id'] 
raw_inventor['citation_name'] = raw_inventor['name_last'].str.upper() +' '+ raw_inventor['name_first'].str.upper()
raw_inventor['citation_inventor_id'] = raw_inventor['inventor_id'] 

raw_inventor = raw_inventor.drop(['name_last'],axis = 1)
raw_inventor = raw_inventor.drop(['name_first'],axis = 1)
raw_inventor = raw_inventor.drop(['application_id'],axis = 1)
raw_inventor = raw_inventor.drop(['inventor_id'],axis = 1)

# merge citation_list and raw_inventor tabels on citation_app
citation_list = pd.merge(citation_list, raw_inventor, on = ['citation_app'])
```

```python
# grouping citation_list by patent_id
citation_list = citation_list.groupby('patent_id').agg(lambda x : list(set(x))).reset_index()
```

## Pickling patent_citation

```python
import pickle

pickle.dump(citation_list, open('patent_citation.pkl', 'wb'))
print("pickled dump finished")
```
