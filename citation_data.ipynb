```python
import pymysql as mysql
import pandas as pd
import numpy as np
import pickle
```

```python
# pickle data import

df_inventors = pd.read_pickle('./samsung_inventors.pkl')
df_inventors[:10]

print(len(set((df_inventors['inventor_id']))))
```

```python
# name_last and name_first Merge & upper string data

df_inventors['name'] = df_inventors['name_last'].str.upper() +' '+ df_inventors['name_first'].str.upper()
df_inventors['organization'] = df_inventors['organization'].str.upper()

# name_last and name_first drop & columns reindexing

df_inventors = df_inventors.drop(['name_last'],axis = 1)
df_inventors = df_inventors.drop(['name_first'],axis = 1)
df_inventors = df_inventors.reindex(columns=['inventor_id', 'name', 'organization', 'application_id', 'number'])
```

```python
#inventor_set = df_inventors.groupby('inventor_id')['name'].apply(lambda x : list(x)).reset_index()
#inventor_set = df_inventors[['inventor_id','name','organization','application_id','number']].groupby('inventor_id').agg("sum")

inventor_set = df_inventors.groupby('inventor_id').agg(lambda x : set(x)).reset_index()
#inventor_set = df_inventors.groupby('inventor_id').agg(lambda x : '~'.join(set(x))).reset_index()
```

```python
inventor_set['name'] = inventor_set['name'].apply(lambda x : list(x)[0])
inventor_set['organization'] = inventor_set['organization'].apply(lambda x : list(x)[0])
```
