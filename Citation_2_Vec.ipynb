## Module 호출

```python
import pandas as pd
import numpy as np
import pickle
```

## Pickle 파일 호출

```python
patent_citation = pd.read_pickle('./patent_citation.pkl')
patent_citation
```

```python
# make some column's data to list
citation_app = patent_citation['citation_app'].values.tolist()
citation_name = patent_citation['citation_name'].values.tolist()
```

## Import Word2vec module 

```python
from gensim.models import Word2Vec

# make ciataion2Vec model 
citation2vec = Word2Vec(citation_app, size = 100, window = 5 , min_count = 1, workers = 4 )
```

```python
citation2vec.most_similar(['2003/20030095250'])
```

```python
# make name2Vec model
name2Vec = Word2Vec(citation_name, size = 100, window = 5 , min_count = 1, workers = 4 )
```

```python
name2Vec.most_similar(['NELSON WILL LEONARD', 'LOIS CAMPBELL ANNIE '] )
```
